```python
import pandas as pd
```

```python
clinical_data = pd.read_csv("data/clinical_data_anonymized.csv")
clinical_data.head()
```

```python
segmentation_data = pd.read_csv("data/segmentation_data_anonymized.csv")
segmentation_data.head()
```

```python

merged_df = pd.merge(clinical_data, segmentation_data, on="name", how="inner")
```

```python
merged_df.head()
```

```python
merged_df = merged_df.drop_duplicates(subset="name", keep="first")
```

```python
merged_df.to_csv("data/final_database_clinical_segmentation.csv")
```
